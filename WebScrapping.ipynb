```python
# install the autoscraper library
from autoscraper import AutoScraper
```

```python
#URL of the page to be scrapped & to hold the information to be scraped
email_url="https://www.cirrusinsight.com/blog/101-sales-email-templates-you-can-use-to-close-more-deals#sales-introduction-email-templates"
wanted_list=["Cold Email Templates","First Follow Up","{Company Name} + {Your Company}","{name}, is the below of any interest to you?"]
```

```python
# generate scraped data.
scraper = AutoScraper()
result = scraper.build(email_url,wanted_list)
data = scraper.get_result_similar(email_url, grouped=True)
print(data)
```

```python
# You can find your keys by the following code
key = list(data.keys())
print(key)
```

```python
# It is similar to training a model and saving it as a file.
scraper.set_rule_aliases({'rule_798j':'Categories', 'rule_j7wh':'Template type', 'rule_p82m':'Subject','rule_3nxj':'Context'})
# keeping the same rules for other queries also
scraper.keep_rules(['rule_798j','rule_j7wh','rule_p82m','rule_3nxj'])
scraper.save('email-temp-search')
```

```python
# we can use this scraper again to scrap similar data from
# specify the path
scraper.load("email-temp-search")
result = scraper.get_result("https://www.cirrusinsight.com/blog/101-sales-email-templates-you-can-use-to-close-more-deals#sales-introduction-email-templates", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('Email_Data.csv')
```
